```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PINNs Architecture
class PINNs(nn.Module):
    def __init__(self):
        super(PINNs, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(1, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 1)
        )

    def forward(self, x):
        return self.model(x)
```

```python
# Physics-Informed Loss Function
def physics_loss(model, V, theta_true, rho=1.0, T=0.5, d=0.1, th=0.2):
    # Enable gradient tracking for input
    V.requires_grad_(True)

    # Forward pass for predictions
    theta_pred = model(V)

    # Calculate R (radius)
    R_pred = d / torch.sin(theta_pred)

    # Calculate resistance (Res)
    term1 = th / R_pred - torch.cos(theta_pred)
    valid_mask = term1.abs() < 1  # Avoid invalid sqrt
    safe_term1 = term1[valid_mask]
    atan_expr = torch.atan(torch.sqrt((1 - safe_term1) / (1 + safe_term1)) * torch.tan(theta_pred[valid_mask] / 2))
    Res_pred = torch.zeros_like(R_pred)
    Res_pred[valid_mask] = (
        rho / (2 * np.pi * R_pred[valid_mask] ** 2) *
        (theta_pred[valid_mask] - safe_term1 / torch.sqrt(1 - safe_term1 ** 2) * atan_expr)
    )

    # Calculate dtheta/dt
    dtheta_dV = torch.autograd.grad(
        outputs=theta_pred, inputs=V,
        grad_outputs=torch.ones_like(theta_pred),
        create_graph=True
    )[0]

    # Voltage dynamics (V_pred)
    V_pred = Res_pred * (4 * np.pi * T * d**2) * (-torch.sin(theta_pred) * dtheta_dV)

    # Compute loss
    physics_error = torch.mean((V - V_pred) ** 2)
    data_error = torch.mean((theta_true - theta_pred) ** 2)

    return physics_error + data_error
```

```python

# Generate synthetic data
def generate_data(n_points=100):
    theta = torch.linspace(0.1, np.pi / 2, n_points).unsqueeze(1)  # Avoid division by zero
    V = 10 * torch.sin(theta)  # Voltage is a function of theta
    return V, theta
```

```python

# Training loop
def train_pinns(model, V, theta, epochs=5000, lr=1e-3):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    loss_history = []

    for epoch in range(epochs):
        optimizer.zero_grad()
        loss = physics_loss(model, V, theta)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())

        if (epoch + 1) % 500 == 0:
            print(f"Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}")

    return loss_history
```

```python
# Generate data
V, theta = generate_data(n_points=100)

# Initialize model
model = PINNs()

# Train the model
loss_history = train_pinns(model, V, theta, epochs=5000, lr=1e-3)
```

```python
 # Plot loss history
plt.figure(figsize=(10, 6))
plt.plot(loss_history, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve for PINNs Training')
plt.legend()
plt.grid(True)
plt.show()
```

```python
 # Test the model on new data points
V_test, theta_test = generate_data(n_points=50)
theta_pred = model(V_test).detach().numpy()
plt.figure(figsize=(10, 6))
plt.plot(theta_test.numpy(), theta_pred, label='Predicted Theta', color='r')
plt.plot(theta_test.numpy(), theta_test.numpy(), label='True Theta', color='b', linestyle='dashed')
plt.xlabel('True Theta')
plt.ylabel('Predicted Theta')
plt.title('Model Prediction vs True Values')
plt.legend()
plt.grid(True)
plt.show()
```
